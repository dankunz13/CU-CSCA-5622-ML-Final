### Explanation of Project

This project aims to classify individuals into one of seven obesity categories (Underweight, Normal, Overweight I, Overweight II, Obese I, Obese II, and Obese III) using 16 predictor variables.  These variables capture physical, dietary, and lifestyle factors that seem to cover a wide range of contributors to obesity. The project utilizes a dataset found at the UC Irvine Machine Learning Repository that includes 3 continuous and 13 categorical variables.  By comparing Multinomial Logistic Regression and Random Forest models, the project evaluates the effectiveness of these machine learning techniques for multiclass classification and uncovers key drivers of obesity.

### Type of Learning/Algorithm

This project uses supervised learning techniques of Multinomial Logistic Regression and Random Forest.  Multinomial Logistic Regression provides a baseline model due to its simiplicity and interpretability, while Random Forest was selected for its ability to handle non-linear relationships and perform feature importance analysis.  Preprocessing and feature engineering will be explored within both algorithms.  Finally, overall prediction accuracy and individual classifier f1-scores, which provide insight into predictive accuracy accross the seven obesity categories, will be used to compare algorithms.

### Type of Task

This is a multiclass classification where the goal is to predict one of seven obesity categories based on physical, activity/lifestyle, and dietary predictors.

### Motivation/Goal of Project

The motivation for this project is to uncover key drivers of obesity and gain insight into the most effective techniques for predicting obesity categories.  Understanding key drivers of obesity and optimizing prediction methods could potentially influence personal health decisions or the focus of public health initiatives.

## Data Source

The data for this project, *Estimation of Obesity Levels Based On Eating Habits and Physical Condition*, (De La Hoz & Tabares, 2019) comes from the UC Irvine Machine Learning Repository. The data and supporting information were initially provided by Palechor & Manotas (2019) in *Dataset for estimation of obesity levels based on eating habits and physical condition in individuals from Colombia, Peru and Mexico*, published in *Data in Brief*. The dataset is provided in a single-table, flat-file format as a CSV file.

The dataset focuses on estimating obesity levels using information from individuals in Mexico, Peru, and Colombia. Specifically, individuals were located in Barranquilla, Colombia; Lima, Peru; and Mexico City, Mexico and between the ages of 14 and 61.  The dataset's features primarily describe the respondents' eating habits and physical conditions.

A unique characteristic of this dataset is that 485 (23%) of the 2,111 records were collected through an anonymous online survey open for 30 days.  The remaining 1,626 (77%) records were generated using Weka and the SMOTE algorithm to balance the distribution of obesity categories. Below are distributions of obesity levels before and after data balancing (Figure below adapted from Fabio Mendoza Palechor, Alexis De la Hoz Manotas (2019)) that show how generated records addressed an imbalance in the resulting obesity classifications. The Obesity classifications benefited most from these generated records.

# Data Cleaning
---
This section describes the original processing performed by the dataset authors and goes through the additional preprocessing steps applied in this project to prepare the data for analysis.

### Original Online Survey Processing

1. **Data Cleaning and Normalization:** Missing and atypical data were deleted and the remaining data was normalized.  The dataset authors reported 485 records resulting from this process, however the exact number of records deleted and the normalization method used are not specified.
2. **Response Variable Calculation:** Obesity Value calculated based on the following formula:$$ Mass Body Index = \frac {Weight}{{Height}*{Height}} $$
3. **Response Variable Categorization:** Obesity Value used to assign record to one of seven Obesity Categories based on data provided by WHO and the Mexican Normativity (*NORMA Oficial Mexicana NOM-008-SSA3-2010, Para el tratamiento integral del sobrepeso y la obesidad*, Diario Oficial (2010)).  The table below shows these seven categories.

| Category | Mass Body Index |
|---|---|
|Underweight|Less than 18.5|
|Normal|18.5 to 24.9|
|Overweight I|25.0 to 26.9|
|Overweight II|27.0 to 29.9|
|Obesity I|30.0 to 34.9|
|Obesity II|35.0 to 39.9|
|Obesity III|Higher than 40|

4. **Balancing Process:** Balancing process using Weka and the SMOTE algorithm utilized to generate records and balance distribution of the obesity category.

An area for improvement would be the initial dataset used. I'm not sure that the dataset with 480 anonymous survey responses and 1626 generated records is representative of a larger population.  Regional or cultural differences may also impact how survey respondents answer questions in different regions.  The balancing that the dataset authors performed didn't seem to fit with some of the asked survey questions, so assumptions needed to be made to categorize some of the variables.  These assumptions may not be valid and could impact performance, but addressing this issue would require starting from and understanding the original survey data.  Another concern with the data is that the authors pre-processing was not well described, so it was difficult to interpret if I was working with raw data, lightly pre-processed data, or heavily pre-processed data.
